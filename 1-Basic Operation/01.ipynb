4. Gabungkan citra bird.bmp(citra A) dan san.bmp (citra B) menggunakan rumus
dimana nilai

```python
img = plt.imread("/content/drive/MyDrive/Colab Notebooks/Citra Digital/tugas citra/bird.bmp")
img1 = plt.imread("/content/drive/MyDrive/Colab Notebooks/Citra Digital/tugas citra/san.bmp")
```

```python
imghasil = img+img1
```

```python
def mean(x,a):
  rata = int(0.8*int(x) + 0.2*int(a))/3
  if rata > 255:
    rata = 255
  if rata < 0:
    rata =0
  return rata

for a in range(256):
  for b in range(256):
    imghasil[a,b] = mean(img[a,b], img1[a,b])
```

```python
plt.imshow(imghasil, cmap="gray")
```

5. Tunjukkan apakah terdapat perbedaan antara citra lena_gray.bmp dan lena_gray2.bmp.

```python
img = plt.imread("/content/drive/MyDrive/Colab Notebooks/Citra Digital/tugas citra/lena_gray.bmp")
```

```python
img1 = np.copy(img)
```

```python
imghasil = img - img1
```

```python
def mean(x,a):
  rata = int(0.8*int(x) - 0.2*int(a))/3
  if rata > 255:
    rata = 255
  if rata < 0:
    rata =0
  return rata

for a in range(256):
  for b in range(256):
    imghasil[a,b] = mean(img[a,b], img1[a,b])
```

```python
plt.imshow(imghasil, cmap="gray")
```

7. Dengan menggunakan bird.bmp dan san.bmp tampilkan hasil operasi logika boolean or terhadap dua citra gray scale tersebut.

```python
img1 = plt.imread("/content/drive/MyDrive/Colab Notebooks/Citra Digital/tugas citra/bird.bmp")
img2 = plt.imread("/content/drive/MyDrive/Colab Notebooks/Citra Digital/tugas citra/san.bmp")
```

```python
imgor = np.copy(img1)
```

```python
for i in range(img1.shape[0]):
  for j in range(img1.shape[1]):
    imgor[i,j] = np.bitwise_or(img1[i,j],img2[i,j])
```

```python
plt.imshow(imgor, cmap="gray")
```
